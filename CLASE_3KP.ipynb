```python
# prompt: crea una tabla de frecuencia y el histograma con su poligono de frecuencia

import matplotlib.pyplot as plt
# @title **Paso 14. Crear tabla de frecuencia y histograma para la variable 'PARCIAL 1'**

# Crear la tabla de frecuencia para 'PARCIAL 1'
tabla_frecuencia_parcial1 = df['PARCIAL 1'].value_counts().reset_index()
tabla_frecuencia_parcial1.columns = ['Valor', 'Frecuencia']
tabla_frecuencia_parcial1 = tabla_frecuencia_parcial1.sort_values(by='Valor')

# Calcular la frecuencia relativa y acumulada
tabla_frecuencia_parcial1['Frecuencia Relativa'] = tabla_frecuencia_parcial1['Frecuencia'] / tabla_frecuencia_parcial1['Frecuencia'].sum()
tabla_frecuencia_parcial1['Frecuencia Acumulada'] = tabla_frecuencia_parcial1['Frecuencia'].cumsum()
tabla_frecuencia_parcial1['Frecuencia Relativa Acumulada'] = tabla_frecuencia_parcial1['Frecuencia Relativa'].cumsum()

print("Tabla de Frecuencia para PARCIAL 1:")
print(tabla_frecuencia_parcial1)

# Crear el histograma y el polígono de frecuencia para 'PARCIAL 1'
plt.figure(figsize=(10, 6))

# Histograma
hist, bins, _ = plt.hist(df['PARCIAL 1'], bins=10, color='skyblue', edgecolor='black', alpha=0.7, label='Histograma')

# Calcular los puntos medios de los intervalos para el polígono de frecuencia
bin_centers = 0.5 * (bins[:-1] + bins[1:])

# Polígono de frecuencia
plt.plot(bin_centers, hist, marker='o', color='red', linestyle='-', label='Polígono de Frecuencia')

plt.title('Histograma y Polígono de Frecuencia para PARCIAL 1')
plt.xlabel('PARCIAL 1')
plt.ylabel('Frecuencia')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
